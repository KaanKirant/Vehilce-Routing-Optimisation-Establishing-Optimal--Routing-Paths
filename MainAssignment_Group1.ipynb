```python
import math #math function for calculating the distance

#Distance calculation function.
def calculate_distance(coordinate_x, coordinate_y, i, j):
  return math.sqrt((abs(coordinate_x[i] - coordinate_x[j])) ** 2 + (abs(coordinate_y[i] - coordinate_y[j])) ** 2) #using the formula for the distance.
```

```python
import numpy as np

dis_car_ij = np.zeros((len(coordinate_x),len(coordinate_x))) #list for distance information
rate_car_ij = np.zeros((len(coordinate_x),len(coordinate_x))) #list for transmission rate information
car_graph = { } # a dictionary to hold information about cars and their neighbours
distance_graph = { } # a dictionary to hold distance information about cars and their neighbours

for i in range(len(coordinate_x)):
    car_graph[car_no[i]] = {} #add each car to the car graph
    distance_graph[car_no[i]] = {} #add each car to the car graph
    for j in range(len(coordinate_x)):
        #calculating the distance between two cars
        dis_car_ij[i][j] = calculate_distance(coordinate_x, coordinate_y, i, j)
        # Here is to calculate the transmission rate according to the distance
        if dis_car_ij[i][j] == 0:
          rate_car_ij[i][j] = 0
        elif dis_car_ij[i][j] < 500:
            rate_car_ij[i][j] = 10
        elif dis_car_ij[i][j] >= 500 and dis_car_ij[i][j] < 1000:
            rate_car_ij[i][j] = 8
        elif dis_car_ij[i][j] >= 1000 and dis_car_ij[i][j] < 2000:
            rate_car_ij[i][j] = 6
        elif dis_car_ij[i][j] >= 2000 and dis_car_ij[i][j] < 3000:
            rate_car_ij[i][j] = 4
        elif dis_car_ij[i][j] >= 3000 and dis_car_ij[i][j] < 4000:
            rate_car_ij[i][j] = 2
        elif dis_car_ij[i][j] >= 4000 and dis_car_ij[i][j] < 6000:
            rate_car_ij[i][j] = 1
        elif dis_car_ij[i][j] >= 6000:
            rate_car_ij[i][j] = 0

        #if the transmission rate is higher than 0 then add this car neigbour connection to both dictionaries.
        if(rate_car_ij[i][j] > 0):
          #print('The distance between Car-' + str(car_no[i]) + ' to Car-' + str(car_no[j]) + ' is = ' + str(dis_car_ij[i][j]))
          #print('The transmission rate between Car-' + str(car_no[i]) + ' to Car-' + str(car_no[j]) + ' is = ' + str(rate_car_ij[i][j]))
          car_graph[car_no[i]][car_no[j]] = rate_car_ij[i][j]
          distance_graph[car_no[i]][car_no[j]] = dis_car_ij[i][j]
```

**Dijkstra Algorithm**

1. Starting Point

  •	Give each automobile a special identification number, then set up its car's latency period and measurement of distance to the base stations 1 and 2.

  •	Assign the originating car's starting length of 0 and all subsequent cars' starting distances of indefinite.
2.	Priority Queue: To monitor the automobiles and their journeys, use a priority queue. Put the      original automobile in the priority queue at the beginning.
3.	The Dijkstra Algorithm

  •	Take the following actions while the priority queue is still filled:

  •	Take the vehicle in the priority line that is the closest to you.

  •	Determine the approximate distance between each neighboring extracted automobile and the original source car.

  •	Adjust the distance and delay values if the estimated distance is shorter than the specified distance as of right now.

  •	If the neighbor's automobile isn't already in the priority queue, add it.
4.	Termination: The approach should be run repeatedly until the priority queue is null.
5.	Routing Path: Following the Dijkstra algorithm's completion, every car's ideal routing route will be identified by taking into account the lowest possible end-to-end latency and data rate of transmission.
6.	Calculating Data Transmission Rate:

  •	Determine the data transmission rate depending on the distance between communication automobiles for each connection in the best routing path.
  •	Find the lowest transmission rate that the route can have.
7.	Compute End-to-End Latency:

  •	Latency of the entire process is the sum of all the latency created by each link across the best possible routing path.
8.	Continue in this manner for every car:

  •	Repeat previous processes for each car to discover the most optimal routing path with the maximum end-to-end data transmission rate and the minimum end-to-end latency.

```python
import sys # The libary sys was imported to represent infinity.
def minimum_Transmission_dijk(route): # Function defination. It takes a list. This list holds the information of the route.
  inf = sys.maxsize # infinty defined
  min = inf # At start minimum is defined as inifity as we are trying to find minimum.
  for car in range(len(route) - 1): #For every car we are going to check the connection between them and the next car which they are connected.
    current_node, next_node = route[car], route[car + 1]
    if next_node not in car_graph[current_node]:
      print("disconnected")
      return 0
    if car == len(route) -1:
      return min
    if car_graph[route[car]][route[car+1]] < min: # If the transmission rate between these two cars are less then the minimum we already found (or defined as infinty) we are changing the minimum.
      min = car_graph[route[car]][route[car+1]] # Changing the minimum to the new minimum

  return min # Return the minimum rate we found.
```

```python
def latency_dijk(route): #latency function to calculate the latency of the route
  latency = 50 # the latency for each connection the route has is 50ms
  if(len(route)-1 == 0): # if there is one car in the route than the latency is 50.
    return 50
  return (len(route)-1) * latency # the latency is the ((number of cars - 1) * 50). The reason behind the minus one is (number of connections = number of cars - 1)
```

**Genetic Algorithm**


The Genetic Algorithm (GA) is a technique that finds vehicle-optimized routes in order to solve the car routing problem. This is a thorough explanation of its process:

1. Fitness Function:
The algorithm begins with the definition of a fitness function. Transmission rate and latency are the two main factors used by this function to assess a route's quality. Additionally, it guarantees that the path stays connected and terminates at either base station (101 or 102).

2. Initialization:
The first step in the process is to generate a preliminary population of possible routes. These paths are produced at random, beginning with a specified starting point. The algorithm ensures that the routes don’t exceed a certain length (maximum of 50 nodes) and extend until reaching either of the base stations.

3. Selection: Tournament Selection
The system selects possible parent routes for the following generation through a tournament selection procedure. A subset of routes is chosen, and the most suitable ones are chosen to serve as parent routes for crossover.

4. Crossover: Single-Point Crossover
There is a single-point crossover operation performed on the selected parent routes. This involves combining parts of two parent routes to create two new routes (offspring).

5. Mutation:
The technique uses a swap mutation to add diversity and further explore the solution space. Within the route, randomly selected nodes switch places, causing the route's structure to change.

6. Termination Condition:
A termination condition is set by the algorithm according to the number of generations. The algorithm stops generating new routes when the maximum number of generations is achieved.

7. Genetic Algorithm Execution:
The population of routes is evolved by the GA through several cycles of crossing, mutation, and selection. The goal of these iterations is to maximize the routes according to fitness.

CONSTRAINTS

1. Connectivity Constraints:
The fitness function implemented in the Genetic Algorithm ensures the validity of routes by checking for connectivity constraints.
These constraints guarantee that each consecutive node in a route is connected, ensuring that cars move only to adjacent nodes in the network.
Any route that violates this constraint by having disconnected segments is considered invalid and receives a fitness score of zero.
2. End Condition with Connectivity: The fitness function verifies that the route terminates at one of the designated base stations (101 or 102) in addition to assessing the transmission rate and latency.
By coming to an end at the specified stations, this condition makes sure that cars stay linked to the network.
3. Route Validity and Connectivity: Routes are assessed for their adherence to network connectivity as well as their efficiency in terms of transmission rate and latency.
This assurance ensures that the developed routes remain consistently connected across the network, which is essential for efficient car routing.
In conclusion, connectivity constraints are essential to the Genetic Algorithm's assessment of fitness. The system looks for more than just transmission rate-based optimum routes.

```python
def latency_genetic(route):#latency function to calculate the latency of the route
  latency = 50 # the latency for each connection the route has is 50ms
  if(len(route)-1 == 0): # if there is one car in the route than the latency is 50.
    return 50
  return (len(route)-1) * latency # the latency is the ((number of cars - 1) * 50). The reason behind the minus one is (number of connections = number of cars - 1)
```

```python
def fitness(route): #Fitness function takes a route and uses minimum_Transmission function to calculate the fitness of this route.
    disconnected = False # This variable holds the information if the route has disconnected during crossover or mutatin.
    weight_of_transmission_rate = 1 # weight of transmission rate
    weight_of_latency = 10 # weight of latency
    for i in range(len(route) - 1): # Itarate through the route
      if route[i] != 101 and route[i] != 102: # If we didnt reach the either of the statitions we can keep itarating.
        current_node, next_node = route[i], route[i + 1] # This two variables hold the information about he current node and the one next to it.
        if next_node not in car_graph[current_node]: # If next node is not connected  to the current node
          disconnected = True # Disconnection is true
    if(disconnected == False and (route[-1] == 101 or route[-1] == 102)): # If route is not disconnected and ends with either base stations this means that this route is valid and we can calculate the fitness.
      fitness =  (weight_of_latency*(1/latency_genetic(route))) + ((minimum_Transmission_genetic(route))* weight_of_transmission_rate)  # Fitness function is: (minimum transmission rate + (1/latency))
      return fitness  # Return the  fitness
    else:  # If there is disconnection or the route does not end with either base stations then the fitness will be 0.
      fitness = 0
      return fitness  # Return the fitness
```

```python
# Generate an initial population of routes
def initial_population(population_size, start_point):
    # Generate initial random routes from starting car to a base station
    initial_routes = []
    # Generate random routes starting from starting car to either Base Station-1 or Base Station-2
    for chr in range(population_size): # Create the amount of population size that was given as a parameter.
        route = [start_point] # Every route starts from starting car.
        while route[-1] not in base_stations: # While last node is not the one of the base stations keep looping.
            if(len(route) > 50):
              break
            neighbors = list(car_graph[route[-1]].keys()) # Get the neighbours of the last node in the route list.
            next_node = random.choice(neighbors) # Randomly choose one of the neighbours for the list.
            route.append(next_node) #Add to the route
        initial_routes.append(route) # Add this newly created route to the initial population
    return initial_routes # return the population that was created.
```

```python
# Tournament selection to choose parents for crossover
def selection(population, k=5):
    # Perform selection to choose parents based on fitness
    selected_parents = []
    for _ in range(len(population)): # We need the same number of chromosome in the next generation as well. This is why we loop the same amount as the population size.
        participants = random.sample(population, k) #Select 5 random routes from the population.
        winner = max(participants, key=lambda route: fitness(route)) #Choose winner by selecting the route with the maximum fitness.
        selected_parents.append(winner) # Add selected parents to a list. These parents will create the next generation using the crossover function.
    return selected_parents
```

```python
# Single-point crossover to create offspring
def crossover(parent1, parent2):
    crossover_point = random.randint(1, min(len(parent1), len(parent2)) - 1) # Choose a single point between the start of parent one and end of parent 2. The reason behind this is, the route are not in the same length.
    child1 = parent1[:crossover_point] + parent2[crossover_point:] # First child takes the first part of parent one and second part of parent two
    child2 = parent2[:crossover_point] + parent1[crossover_point:] # Second child takes the second part of parent one and first part of parent two
    return child1, child2

# Swap mutation to introduce variability
def mutation(route):
    if random.random() < mutation_rate: # There is a percent of chance that there will be a mutation. This rate was choosen 0.1. If random number we generate is less than that then there is a mutation.
        if(len(route) < 3):
          return route
        idx1, idx2 = random.sample(range(1, len(route)), 2) #Select two random nodes in the route.
        route[idx1], route[idx2] = route[idx2], route[idx1] #Change the position of the two nodes.
    return route # return the changed route.
```

```python
# Termination condition based on the number of generations or a specific fitness level
def should_terminate(generation, max_generations):
    return generation >= max_generations # If we reached the maximum generation stop.

def genetic_algorithm(population_size, max_generations, start_point):
    population = initial_population(population_size, start_point) #Create the initial population using the function we created at the start
    for generation in range(max_generations): # Itarate through this loop in the range of maximum generation.
        parents = selection(population) # Choose parents using the selection function we created at the start
        offspring = [] # The childs that are about to be create will be hold in this list.
        for i in range(0, len(parents), 2): # Itarate two by two selection two parents. This loop  will itarate maximum generation / 2 times. But it will create two childs so the population size will not change
          if(i == len(parents)-1): #If we reached last parent stop. (This only happens if we gave population size that is an odd number. If it is even number the last parent will never be reached. Because funciton itartes two by two.)
            break
          child1, child2 = crossover(parents[i], parents[i + 1]) # Use the crossover function tto create the new two childs.
          child1 = mutation(child1) # Mutate the childs (Only if the chance of mutation was reached.)
          child2 = mutation(child2) # Mutate the childs (Only if the chance of mutation was reached.)
          offspring.extend([child1, child2]) # Add the two newly created childs to the offspring list.
        population = offspring # Our new population is the offsprings.
        if should_terminate(generation, max_generations): # If we reached the maximum generation stop
            break
    best_route = max(population, key=lambda route: fitness(route)) # Get the result from using the function above. From the last population select the node with the maximum fitness.
    return best_route

# Execute the genetic algorithm
population_size = 300
mutation_rate = 0.1
max_generations = 100
```

**Particle Swarm Optimization**

While implementing particle swarm optimization a different approach was taken. The Cars were created as a class and have their own variables such as x coordinate and y coordinate and car number. The particles were created in a similar way where they variables such as path, path cost and holds the information about the global best and personal best. The object-oriented approach will be beneficial for particle swarm optimization as particles have their unique identity and have their own values.

To initialize the particles a similar function was used for genetic algorithms. Starting from starting car that was randomly chosen to create different routes until it reaches one of the base stations. Cost function is also like genetic algorithms. The cost was defined as the sum of the transmission rate and the reciprocal of latency. The weight values were chosen the same as genetic algorithms for comparison purposes. The reason 1 divided by latency was chosen is to minimize the latency. The goal is to maximize the transmission rate and minimize the latency.

The change from genetic algorithm starts in the update function where genetic algorithm used crossover to create the next population, particle swarm algorithm updates the current population using a velocity. The velocity for this case is the changes will be made to the particle route based on personal best and global best. The goal is to get close to current best solution in the population. The update changes some cars in the route to be more similar to better solutions.

In summary, the object-oriented approach, along with subtle changes in the initialization and update procedures, separates the PSO implementation from its genetic algorithm equivalent. The emphasis on personalized entities, as well as the use of personal and global best solutions, contribute to the algorithm's capacity to iteratively improve upon existing solutions, making it an effective optimization strategy for problems with distinct identities and dynamic variables.

```python
def path_cost(route):
    disconnected = False # This variable holds the information if the route has disconnected.
    weight_of_transmission = 1 # weight of transmission rate
    weight_of_latency = 10 # weight of latency
    for i in range(len(route) - 1): # Itarate through the route
      if route[i].no != 101 and route[i].no != 102: # If we didnt reach the either of the statitions we can keep itarating.
        current_node, next_node = route[i].no, route[i + 1].no # This two variables hold the information about he current node and the one next to it.
        if next_node not in car_graph[current_node]: # If next node is not connected  to the current node
          disconnected = True # Disconnection is true
    if(disconnected == False and (route[-1].no == 101 or route[-1].no == 102)): # If route is not disconnected and ends with either base stations this means that this route is valid and we can calculate the fitness.
      fitness =  (weight_of_latency*(1/latency_pso(route))) + ((minimum_Transmission_pso(route))* weight_of_transmission)  # Fitness function is: (minimum transmission rate + (1/latency))
      return fitness  # Return the minimum transmission rate as fitness
    else:  # If there is disconnection or the route does not end with either base stations then the fitness will be 0.
      fitness = 0
      return fitness  # Return the minimum transmission rate as fitness
```

```python
import sys # The libary sys was imported to represent infinity.
def minimum_Transmission_pso(route): # Function defination. It takes a list. This list holds the information of the route.
  inf = sys.maxsize # infinty defined
  min = inf # At start minimum is defined as inifity as we are trying to find minimum.
  for car in range(len(route) - 1): #For every car we are going to check the connection between them and the next car which they are connected.
    current_node, next_node = route[car].no, route[car + 1].no # Changes was made from the original function to fit the class approach that was used for PSO.
    if next_node not in car_graph[current_node]:
      return 0
    if car == len(route) -1:
      return min
    if car_graph[route[car].no][route[car+1].no] < min: # If the transmission rate between these two cars are less then the minimum we already found (or defined as infinty) we are changing the minimum.
      min = car_graph[route[car].no][route[car+1].no] # Changing the minimum to the new minimum

  return min # Return the minimum rate we found.
```

```python
class Particle: # Particle class.
    def __init__(self, route, cost=None):
        # Current route
        self.route = route # Each particle have a route
        # Personal best route
        self.pbest = route # A personal best for particle
        # Current cost
        self.current_cost = cost if cost else self.path_cost() # Cost of the particle route
        # Personal best cost
        self.pbest_cost = cost if cost else self.path_cost() # Cost of the personal best route
        self.velocity = [] # Velocity holds the information about the changes that is going to be made to the particle route

    def clear_velocity(self): # Clears the velocity
        self.velocity.clear()

    def update_costs_and_pbest(self): # After the changes change the cost information
        self.current_cost = self.path_cost()
        if self.current_cost < self.pbest_cost:
            self.pbest = self.route
            self.pbest_cost = self.current_cost

    def path_cost(self): # Calculate the cost
        return path_cost(self.route)
```

```python
class PSO: # Main class that runs the PSO algorithm

    # Initialise the Optimisation object with the values supplied
    def __init__(self, iterations, population_size, starting_car, base_stations, gbest_probability=1.0, pbest_probability=1.0, cars=None):
        self.cars = cars
        # Global best route
        self.gbest = None
        self.gcost_iter = []
        self.iterations = iterations
        self.population_size = population_size
        self.particles = []
        self.gbest_probability = gbest_probability
        self.pbest_probability = pbest_probability
        self.starting_car = starting_car
        self.base_stations = base_stations

        solutions = self.initial_population()
        # Set up a list of particle objects for each solution (route) in the initial population
        self.particles = [Particle(route=solution) for solution in solutions]

    # Construct a completely random route
    def random_route(self):
        route = [cars[self.starting_car-1]] # Every route starts from starting car.
        while route[-1].no not in self.base_stations: # While last node is not the one of the base stations keep looping.
          if(len(route) > 50):
            break
          neighbors = list(car_graph[route[-1].no].keys()) # Get the neighbours of the last node in the route list.
          next_node = random.choice(neighbors) # Randomly choose one of the neighbours for the list.
          route.append(cars[next_node - 1]) #Add to the route
        return route

    # Construct a population from completely random routes created
    def initial_population(self):
        random_population = [self.random_route() for _ in range(self.population_size)]
        return [*random_population]

    # This performs the main optimisation
    def run(self):
        # global best is the minimum of the particle costs
        self.gbest = max(self.particles, key=lambda p: p.pbest_cost)

        for t in range(self.iterations):
            self.gbest = max(self.particles, key=lambda p: p.pbest_cost)
            # Add best cost to list
            self.gcost_iter.append(self.gbest.pbest_cost)

            for particle in self.particles:
                particle.clear_velocity()
                temp_velocity = []
                gbest = self.gbest.pbest[:]
                new_route = particle.route[:]

                # Set up changes based on the particles personal best position/route
                for i in range(len(particle.route)-1):
                    if i == len(new_route) or i == len(particle.pbest):
                      random_node_in_pbest = random.randint(1,len(particle.pbest)-1)
                      random_node_in_newroute = random.randint(1,len(new_route)-1)
                      swap = (random_node_in_newroute, random_node_in_pbest, self.pbest_probability,0) # 1 index to change on new particle, 2 index of that car no in the route gbest, 3 probabilty
                      temp_velocity.append(swap)
                      break
                    if new_route[i] != particle.pbest[i]:
                        swap = (i, i, self.pbest_probability,0)
                        temp_velocity.append(swap)
                        break


                # Set up changes based on the swarms global best position/route
                for i in range(len(particle.route)-1):
                    if i == len(new_route) or i == len(gbest):
                      random_node_in_gbest = random.randint(1,len(gbest)-1)
                      random_node_in_newroute = random.randint(1,len(new_route)-1)
                      swap = (random_node_in_newroute, random_node_in_gbest, self.gbest_probability,1) # 1 index to change on new particle, 2 index of that car no in the route gbest, 3 probabilty
                      temp_velocity.append(swap)
                      break
                    if new_route[i] != gbest[i]:
                        swap = (i, i, self.gbest_probability,1) # 1 index to change on new particle, 2 index of that car no in the route gbest, 3 probabilty
                        temp_velocity.append(swap)
                        break

                # temp_velocity is a list of swaps to be applied to the new route
                particle.velocity = temp_velocity

                # Apply the velocity changes to new route
                # Velocity change means swapping the order of some of the cities
                for swap in temp_velocity:
                  if swap[3]:
                    if random.random() <= swap[2]:
                        new_route[swap[0]] = gbest[swap[1]]
                  else:
                    if random.random() <= swap[2]:
                        new_route[swap[0]] = particle.pbest[swap[1]]

                # Update the particle
                particle.route = new_route
                particle.update_costs_and_pbest()

                first_base = len(particle.route) - 1
                for car in particle.route:
                  if car.no == 101 or car.no == 102:
                    if particle.route.index(car) < first_base:
                      first_base = particle.route.index(car)

                if particle.route[first_base].no == 101 or particle.route[first_base].no == 102:
                  for i in range(len(particle.route)-1, 0, -1):
                    if i != first_base:
                      particle.route.pop(i)
                    else:
                      break
```

**Experiment Analysis**

Experimenting with certain variables in the algorithm can change the solution that was reached by the algorithms. To get better results some different values was used for the following variables:
1.	Weights of the objectives
The weight approach was used for this Mult objective problem. Changing and tuning these variables has a great effect on the results. During the creation of the algorithm same weights were used. To get better results changes such as increasing the weight of latency was implemented. Until reaching the optimal solution the changes was made and best weights was one for transmission rate and ten for latency.
2.	Population size and iteration
Both genetic algorithm and particle swarm optimization have variables such as population size and iteration. Making changes on these variables can create a better solution but the cost of this was the need for more computational power. Increasing these values effects how long it takes to finalize the results. In an environment where there is better computational power can create better solution. For this problem the values that was chosen was 300 for population size and 100 for iteration.
3.	Global best probability and personal best probability
For particle swarm algorithm two values were experimented to find the best solution. The global best probability and personal best probability are the probability of the swap function running. These can be considered as social learning rate and cognitive learning rate. Changing these values had some changes of how the particles evolved.
4.	Mutation Rate
For genetic algorithms the mutation rate is similar to particle swarms learning rate. The probability of the mutation function running is defined by this variable. Experimenting with this variable had effect on how much change was impacting the population. The optimal value was selected as 0.1.
5.	Population / Particle Initialization
There was some experimenting done on the initialization part of the genetic algorithm and particle swarm algorithm. The rule of repetition was implemented in the first stages of the program. But this had a bad effect on results and changed later. The repetition rule was to check if the route had same car twice. Having same car twice is not good for latency and this rule was implemented at first to make sure the latency was optimized properly. But the results was not good enough for the problem at hand so changes was made and this rule was deleted.
